```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 400)
y = x
plt.figure(figsize=(5,5)) 
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of Identity')

plt.grid(True)
plt.show()
```

```python

def step_function(x):
    return np.where(x >= 0, 1, 0)

x = np.linspace(-5, 5, 1000)

y = step_function(x)

plt.figure(figsize=(5,3))

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Binary Step Function')

plt.grid(True)
plt.show()
```

```python


def logistic_function(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 400)

y = logistic_function(x)

plt.figure(figsize=(7,2))

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Logistic Function a.k.a soft step')

plt.grid(True)
plt.show()
```

```python


def tanh_function(x):
    return np.tanh(x)

x = np.linspace(-10, 10, 400)

y = tanh_function(x)

plt.figure(figsize=(5,3))

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('tanh Function')


plt.grid(True)
plt.show()
```

```python


x = np.linspace(-10, 10, 400)

y = np.arctan(x)

plt.figure(figsize=(7,3))

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('arctan Function')

plt.grid(True)
plt.show()
```

```python


def relu_function(x):
    return np.maximum(0, x)

x = np.linspace(-10, 10, 400)

y = relu_function(x)

plt.figure(figsize=(5,3))

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('ReLU Function')

plt.grid(True)
plt.show()
```

```python


def prelu_function(x, alpha):
    return np.where(x >= 0, x, alpha * x)

x = np.linspace(-10, 10, 400)

alpha = 0.1

y = prelu_function(x, alpha)

plt.figure(figsize=(5,2))

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Parametric ReLU (PReLU) Function')


plt.grid(True)
plt.show()
```

```python


def elu_function(x, alpha):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

x = np.linspace(-5, 5, 400)

alpha = 1.0

y = elu_function(x, alpha)

plt.figure(figsize=(5,2))

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential Linear Unit (ELU) Function')


plt.grid(True)
plt.show()
```

```python

def softplus_function(x):
    return np.log(1 + np.exp(x))

x = np.linspace(-5, 5, 400)

y = softplus_function(x)

plt.figure(figsize=(5,2))

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Softplus Function')

plt.grid(True)
plt.show()
```
